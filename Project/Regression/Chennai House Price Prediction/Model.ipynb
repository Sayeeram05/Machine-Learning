```python
X.head()
```

```python
y = Df.loc[:,"SALES_PRICE"]
```

```python
X_Train,X_Test,y_Train,y_Test = train_test_split(X,y,shuffle=True,test_size=0.25)
```

```python
X_Train.info()
```

```python
plt.figure(figsize=(20,14))
sns.heatmap(X.corr(),annot=True,cmap=sns.color_palette("flare"))
plt.title('Heatmap of the Data')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_Train,y_Train)
```

```python
Prediction = LR.predict(X_Test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import math
```

```python
print("Mean Absolute Error : ",mean_absolute_error(y_Test,Prediction))
print("Mean Square Error : ",mean_squared_error(y_Test,Prediction))
print("Root Mean Square Error : ",math.sqrt(mean_squared_error(y_Test,Prediction)))
```
